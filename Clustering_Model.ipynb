```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ChurnData.csv')
df.head()
```

```python
corr=df.corr()
corr_target=abs(corr['churn'])
relavant_feature=corr_target[corr_target>0.3]
relavant_feature
```

```python
inp_data=pd.DataFrame()
inp_data
```

```python
inp_data = df[['tenure','employ','callcard','loglong']]
inp_data
```

```python
from sklearn.preprocessing import MinMaxScaler
mmc=MinMaxScaler()
inp_data=mmc.fit_transform(inp_data)
inp_data
```

```python
inp_data_nor=pd.DataFrame(inp_data,columns=['tenure','employ','callcard','loglong'])
inp_data_nor
```

```python
out_data = [df['churn']]
out_data
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 2,max_iter=50000)
kmeans
```

```python
kModel = kmeans.fit(inp_data_nor)
```

```python
kModel.labels_
```

```python
pd.crosstab(out_data, kModel.labels_)
```

Work on heart dataset

```python
df=pd.read_csv("heart.csv")
df
```

```python
X=df.drop('target',1)
Y=df.target
```

```python
cor=df.corr()
cor_target=abs(cor['target'])
relevant_feature=cor_target[cor_target>=0.4]
relevant_feature
```

```python
X_new=df[['cp','thalach','exang','oldpeak']]
featureset=X_new
from sklearn.preprocessing import MinMaxScaler
x=featureset.values
x
```

```python
mmc=MinMaxScaler()
feature_mtx=mmc.fit_transform(x)
X_new=feature_mtx
X_new
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2, max_iter=50000)
kmodel=kmeans.fit(X_new)
```

```python
pd.crosstab(Y,kmodel.labels_)
```
